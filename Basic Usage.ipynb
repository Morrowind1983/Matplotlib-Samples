```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 50)
y1 = 1 / (1 + np.exp(-x))
y2 = np.arctan(x)
y3 = [1 if a>0 else 0 for a in x]
```

```python
plt.figure(num=1, figsize=(10,1))
plt.plot(x, y1)

plt.figure(num=2, figsize=(10,2.7468))
plt.plot(x, y2)

plt.figure(num=3, figsize=(10,1))
plt.plot(x, y1)
plt.plot(x, y3)

plt.show()
```
