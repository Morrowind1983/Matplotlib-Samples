```python
import matplotlib.pyplot as plt
import numpy as np
```

### 一个简单的例子

figure相当于画板  
每个figure都可以包含一个或多个axes（轴域）  
创建包含axes的figure，最简单的方法是用pyplot.subplots

```python
fig, ax = plt.subplots()  # 创建一个只有一个axes的figure
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # 在axes上绘制Plot
plt.show()
```

### Figure的组成部分

![avatar](https://matplotlib.org/_images/anatomy.png)

#### Figure

```python
fig = plt.figure()  # 创建一个不包含任何axes的空figure
fig, ax = plt.subplots()  # 创建一个只有一个axes的figure
fig, axs = plt.subplots(2, 2)  # 创建一个包含2x2axes网格的figure
plt.show()
```

#### Axes

一个axes包含2个（或3个）axis  
每个axes有一个title（标题），一个x-label，一个y-label

#### Axis

数轴一样的对象。用于设置图像范围，生成ticks（刻度），ticklabels（刻度标签）。

#### Artist

所有在figure中能看到的东西都是artist。大多数artists绑定到一个axes中。

### 面向对象的接口 和 pyplot接口

有两种方式来使用Matplotlib:

- 显式地创建figures和axes对象，并调用figure和axes对象的方法(面向对象的风格)
- 依赖pyplot自动创建和管理figures和axes，并使用pyplot的函数来绘制

```python
# 面向对象的风格
x = np.linspace(0, 2, 100)

# 即使面向对象风格下，依然用pyplot.figure来创建figure
fig, ax = plt.subplots()  # 创建一个只有一个axes的figure
ax.plot(x, x, label='linear')  # 在axes上绘制plot
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')  # 在axes上添加一个x-label
ax.set_ylabel('y label')  # 在axes上添加一个y-label
ax.set_title("Simple Plot")  # 在axes上添加一个title
ax.legend()  # 在axes上添加一个图例

plt.show()
```

```python
# pyplot函数的风格
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # 在（隐藏的）axes上绘制plot
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()

plt.show()
```

建议在交互式绘制时（如Jupyter notebook）使用pyplot函数，在非交互式绘制时（如一些大工程的复用脚本中）使用面向对象的风格

如果一种图像需要用不同的数据多次绘制，最好写一个函数来绘制：

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

像这样调用上面的函数

```python
data1, data2, data3, data4 = np.random.randn(4, 100)

fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})

plt.show()
```
